# ANALYSIST ON BANKNOTES SECURITY FEATURES

## Banknotes 
is a piece of paper money, constituting a central bank's promissory note to pay a stated sum to the bearer on demand. Banknotes were originally issued by commercial banks, which were legally required to redeem the notes for legal tender (usually gold or silver coin) when presented to the chief cashier of the originating bank. [source](https://en.wikipedia.org/wiki/Banknote) To prevent counterfeiting, it's important to secured Banknotes with the best security features.
<br>
<br>
Each country has their own concern on choosing the type of security features for its banknotes. Sometimes, the higher the denominations, the better security features on it. But, some country choose to standardize their banknotes security features, without concerning the denomination.
<br>
<br>
Here we try to analyze the security features of banknotes from 115 countries. <br>

### Top 10 Countries, with most Security Features
Here we try to visualize top 10 Countries with most Security Features, with **Circular Barplot**.

```python
# sort the data

df_sorted = df.nlargest(10, "Total")
```

```python
# set the angle

ANGLES = np.linspace(0.05, 2 * np.pi - 0.05, len(df_sorted), endpoint=False)
```

```python
# define Total

TOTAL = df_sorted["Total"].values
```

```python
# define Country

COUNTRY = df_sorted["Country"].values
```

```python
# Prepare the graph (1)

GREY12 = "#1f1f1f"

# Set default font to Bell MT
plt.rcParams.update({"font.family": "Bell MT"})

# Set default font color to GREY12
plt.rcParams["text.color"] = GREY12

# The minus glyph is not available in Bell MT
# This disables it, and uses a hyphen
plt.rc("axes", unicode_minus=False)

# Colors
COLORS = ["#231942","#5e548e","#9f86c0","#e0b1cb"]

# Colormap
cmap = mpl.colors.LinearSegmentedColormap.from_list("my color", COLORS, N=256)

# Normalizer
norm = mpl.colors.Normalize(vmin=TOTAL.min(), vmax=TOTAL.max())

# Normalized colors. Each number of tracks is mapped to a color in the 
# color scale 'cmap'
COLORS = cmap(norm(TOTAL))
```

```python
# Prepare the graph (2) - Customize

# Initialize layout in polar coordinates
fig, ax = plt.subplots(figsize=(9, 12.6), subplot_kw={"projection": "polar"})

# Set background color to white, both axis and figure.
fig.patch.set_facecolor("white")
ax.set_facecolor("white")

#set axis
ax.set_ylim(-5, 25)

# Add bars to represent the cumulative track lengths
ax.bar(ANGLES, TOTAL, color=COLORS, alpha=0.9, width=0.52, zorder=10)

# Add dashed vertical lines. These are just references
ax.vlines(ANGLES, 0, 3000, color=GREY12, ls=(0, (4, 4)), zorder=11)

COUNTRY = ["\n".join(wrap(r, 5, break_long_words=False)) for r in COUNTRY]
COUNTRY

# Set the labels
ax.set_xticks(ANGLES)
ax.set_xticklabels(COUNTRY, size=12);

# Remove lines for polar axis (x)
ax.xaxis.grid(False)

# Remove spines
ax.spines["start"].set_color("none")
ax.spines["polar"].set_color("none")

# Add extra space for the labels
XTICKS = ax.xaxis.get_major_ticks()
for tick in XTICKS:
    tick.set_pad(10)

# Add legend
# Add extra space for legend
fig.subplots_adjust(bottom=0.175)

# Create an inset axes.
cbaxes = inset_axes(
    ax, 
    width="100%", 
    height="100%", 
    loc="center",
    bbox_to_anchor=(0.325, 0.1, 0.35, 0.01),
    bbox_transform=fig.transFigure # Note it uses the figure.
) 

# Create a new norm, which is discrete
bounds = [0, 10, 15, 20, 25]
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

# Create the colorbar
cb = fig.colorbar(
    ScalarMappable(norm=norm, cmap=cmap), 
    cax=cbaxes, # Use the inset_axes created above
    orientation = "horizontal",
    ticks=[0, 10, 15, 20, 25]
)

# Remove the outline of the colorbar
cb.outline.set_visible(False)

# Remove tick marks
cb.ax.xaxis.set_tick_params(size=0)

# Set legend label and move it to the top (instead of default bottom)
cb.set_label("Amount of Security Features", size=12, labelpad=-40)

# Add annotations 
# Make some room for the title and subtitle above.
fig.subplots_adjust(top=0.8)

# Define title, subtitle, and caption
title = "\nCountries with Most Security Features on Banknotes"
subtitle = "\n".join([
    "This is the top 10 countries with most security features printed on their Banknotes."
])
caption = "(fig. 1) Data Visualisation of Countries with the most Security Features on Banknotes"

# And finally, add them to the plot.
fig.text(0.1, 0.93, title, fontsize=25, weight="bold", ha="left", va="baseline")
fig.text(0.1, 0.9, subtitle, fontsize=14, ha="left", va="top")
fig.text(0.5, 0.025, caption, fontsize=10, ha="center", va="baseline")
```

## The Most Counterfeited Banknotes 

Counterfeit money is currency produced without the legal sanction of a state or government, usually in a deliberate attempt to imitate that currency and so as to deceive its recipient. Producing or using counterfeit money is a form of fraud or forgery, and is illegal. Some of the ill-effects that counterfeit money has on society include, a reduction in the value of real money; and an increase in prices (inflation) due to more money getting circulated in the economy—an unauthorized artificial increase in the money supply; a decrease in the acceptability of paper money; and losses, when traders are not reimbursed for counterfeit money detected by banks, even if it is confiscated. [source](https://en.wikipedia.org/wiki/Counterfeit_money) <br>
